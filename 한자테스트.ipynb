```python
import os
import time
import datetime
import numpy as np
from IPython.display import clear_output
from hanja import *

dir_db = os.path.join('./한자표')

# 지금까지 익힌 한자 급수를 입력하세요.
levels = ['8급', '7급', '6급', '준5급']

# 시험 문제가 출제될 한자 급수를 입력하세요
tlevels = ['8급'] #, '7급', '6급', '준5급']

# 한자 찾기 보기 수
q2_examples = 4

hanjas, level_hanjas = read_level_hanjas(dir_db, levels)
thanjas, tlevel_hanjas = read_level_hanjas(dir_db, tlevels)
```

# 1. 한자의 뜻(훈) 과 소리(음) 맞추기

```python
clear_output(wait=True)
_thanjas = np.random.permutation(thanjas)
num_correct = 0
for i, hanja in enumerate(_thanjas):
    correct = question1(hanja, num=i+1)
    if correct:
        num_correct += 1

        
print('******** 시험이 종료되었습니다 ********')
print('총 %d 문제중 %d 문제를 맞추셨습니다.' % (len(thanjas), num_correct))
print('********** %.1f 점 획득 ***********' % (num_correct / len(thanjas) * 100))
```
